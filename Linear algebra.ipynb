```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
sns.set()
```

```python
#example plot
plt.quiver(0,0,5,3)
plt.show()
```

```python
plt.quiver(0,0,5,3,scale_units = 'xy',angles = 'xy',scale = 1,color = 'b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#addition in vector
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
sum = vector_1 + vector_2
print(sum)
```

```python
#visualaization in graph
plt.quiver(0,0,2,3,scale_units = 'xy',angles = 'xy',scale = 1,color = 'b')
plt.quiver(0,0,3,-2,scale_units = 'xy',angles = 'xy',scale = 1,color = 'r')
plt.quiver(0,0,5,1,scale_units = 'xy',angles = 'xy',scale = 1,color = 'g')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#subtration in vector
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])

sub = vector_1 - vector_2
print(sub)
```

```python
#visualaization in graph
plt.quiver(0,0,2,3,scale_units = 'xy',angles = 'xy',scale = 1,color = 'b')
plt.quiver(0,0,3,-2,scale_units = 'xy',angles = 'xy',scale = 1,color = 'r')
plt.quiver(0,0,-1,5,scale_units = 'xy',angles = 'xy',scale = 1,color = 'g')
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#multiply in vector useing scalar
vector_1 = np.asarray([0,0,2,3])
size = 2
mul = size * vector_1
print(mul)

vector_2= np.asarray([0,0,2,3])
size = -0.5
mul = size * vector_2
print(mul)
```

```python
#visualaization in graph
plt.quiver(0,0,2,3,scale_units = 'xy',angles = 'xy',scale = 1,color = 'b')
plt.quiver(0,0,4,6,scale_units = 'xy',angles = 'xy',scale = 1,alpha=0.6,color = 'r')
plt.xlabel('X Axis Label')
plt.xlabel('its directly increace a it into double')
plt.ylabel('Y Axis Label')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#visualaization in graph
plt.quiver(0,0,2,3,scale_units = 'xy',angles = 'xy',scale = 1,color = 'b')
plt.quiver(-0,-0,-1,-0.5,scale_units = 'xy',angles = 'xy',scale = 1,alpha=0.6,color = 'r')
plt.xlabel('X Axis Label')
plt.xlabel('its directly decreace  it into double')
plt.ylabel('Y Axis Label')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
